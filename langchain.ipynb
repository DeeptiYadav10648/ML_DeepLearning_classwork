```python
import os
from langchain.chat_models import init_chat_model
os.environ["OPENROUTER_API_KEY"] = os.getenv("OPENROUTER_API_KEY")  #get with .env file

model = init_chat_model (
    "auto",
    model_provider="openrouter",
    temperature=0.2,  #temperature controls randomness and creativty.
    # max_tokens=204,
    # max_retries=3,
)
```

```python
#TEMPERATURE CONTROLS CREATIVITY AND RANDOMNESS OF THE MODEL'S OUTPUT.
#  A LOWER TEMPERATURE WILL MAKE THE MODEL MORE DETERMINISTIC AND FOCUSED, WHILE A HIGHER TEMPERATURE WILL MAKE IT MORE CREATIVE AND DIVERSE.
#max_tokens controls the maximum number of tokens that the model can generate in response to a given input.
#  It limits the length of the output generated by the model, ensuring that it does not exceed a specified token count. 
# This parameter is useful for managing the verbosity of the model's responses and preventing excessively long outputs.
#max_retries controls the number of times the model will attempt to generate a response if it encounters an error or fails to produce a valid output.
#  It specifies the maximum number of retry attempts before giving up and returning an error.
```

```python
# in gen ai we use 3 functions:
# 1. invoke: to get the response from the model.
# 2. stream: to get the response from the model in a streaming fashion.
#3. batch: to get the response from the model for a batch of inputs.
```

```python
response = model.invoke(
    [
        ("user", "What is the capital of India?")
    ]
)
print(response)
print(response.content)
print(response.text)
```

```python
from langchain_core.prompts import ChatPromptTemplate
import langchain_core
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
```

```python
prompt = ChatPromptTemplate.from_template("You are a developer. Answer the following topic in a witty manner: {question}   ")
```

```python
#role is scientist
result = model.invoke(
    prompt.format_prompt(question="BigBang Theory?").to_messages()
)

print(result.content)
```

```python
#role is developer
result = model.invoke(
    prompt.format_prompt(question="BigBang Theory?").to_messages()
)

print(result.content)
```
